### Fast RCNN

R-CNN modelinde 2000 farkl aday blge karlmas ve bu 2000 blge iin, 2000 farkl CNN a kullanlmas eitim sreci anlamnda ok byk maliyete sebep olmakta. Bunun iin Fast R-CNN mimarileri 2000 CNN modelinden kurtulup sadece tek bir model kullanmaktadr.

Fast R-CNNde odaklanlmas gereken en byk gelitirme R-CNNde kullanlan CNN, SVM ve Regressoru kombinlemesidir. Bu kombinleme ile muazzam bir performans stnl elde etmektedir.

Fast R-CNNde:

1- Tm grsel CNN ile ilenir ve feature mapler (znitelik haritalar) karlr

2- Her blge nerisi iin gerekli znitelikler toplanr. Bunlara region proposal feature map denir. Ardndan bu haritalar max pooling (maksimum havuzlama) ile belirli boyutlara indirgenir. Bu havuzlama katmanna ise RoI (Region of Interest) pooling layer ad verilir.

3- Ardndan bu znitelik haritalar tek boyutlu vektr haline indirgenir.

4- Bu vektrler sinir ana sokulur ve softmax ile blgedeki objeye ait snf, bounding box regressor ile de objenin snrlayc kutusu belirlenir.
